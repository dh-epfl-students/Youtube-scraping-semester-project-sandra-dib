```python
# create a csv file
with open('unesco_sites.csv', 'w') as f:

    for link in links1:
        #print(links1)
        response = requests.get(
        url='https://en.wikipedia.org'+link
        )

        soup = BeautifulSoup(response.text, 'html.parser')
        table=soup.find('table',{'class':"wikitable"})
        df=pandas.read_html(str(table))

        # convert list to dataframe
        df=pandas.DataFrame(df[0])
        #newdf = df.dropna(axis = 0, how ='any')
        #newdf = df.replace('#', '', encoding='ufeff')
        newdf = df.fillna('')
        #print(df.head())

        # drop the unwanted column
        if 'Site' in newdf.columns and 'Location' in newdf.columns:
            data = newdf[['Site', 'Location']]
            #print(link,' retrieved: ', data.columns)
            #print(data.head())
            #print()
            # add information to the csv file
        
            
            for index, value in data.iterrows():
                #print(index, value['Site'], value['Location'])
                #value.create_from_csv_row(row.encode('ascii', 'ignore'))
                location_cleaned = unicodedata.normalize('NFKD', value['Location']).encode('ascii','ignore').decode('ascii')
                f.write("%s;%s;%s\n" % (link, value['Site'], location_cleaned))

                
                
        #data = df.drop(["Rank", "Population(2001)", "Population(2011)[3]"], axis=1)
        #data = data.rename(columns={"State or union territory": "State"})
        #print(data.head())
```
