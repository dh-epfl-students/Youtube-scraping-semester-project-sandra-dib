```python
html = urllib.request.urlopen("https://www.youtube.com/results?search_query=Aksum+drone")
video_ids = re.findall(r"watch\?v=(\S{11})", html.read().decode())
print(video_ids)
```

```python
import virtualenv
from bs4 import BeautifulSoup
import requests
import pafy
```

```python
HEADER = {
    "User-Agent": 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
}
# turn HTML content into structured tree
for video_id in video_ids:
    response = requests.get("https://www.youtube.com/watch?v=" + video_id , headers=HEADER)
    soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup
```

```python
# Extract title
titleSoupMeta = soup.find("meta", property="og:title")
videoTitle = titleSoupMeta["content"] if titleSoupMeta else "NotFound"

print(videoTitle)
```

```python
import youtube_dl

ydl = youtube_dl.YoutubeDL({'outtmpl': '%(id)s.%(ext)s'})

for video_id in video_ids:
    with ydl:
        result = ydl.extract_info(
            "https://www.youtube.com/watch?v=" + video_id,
            download=False # We just want to extract the info
        )

    if 'entries' in result:
        # Can be a playlist or a list of videos
        video = result['entries'][0]
    
    else:
        # Just a video
        video = result

    print(video["title"])
```
